# 多層感知器實作

## 載入套件

```python
import tensorflow as tf
import input_data

mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

## 訓練資料

```python
batch_xs, batch_ys = mnist.train.next_batch(5)
```

## 建立Computational Graph

```python
x_ = tf.placeholder(tf.float32, [None, 784], name="x_")
y_ = tf.placeholder(tf.float32, [None, 10], name="y_")

W1 = tf.Variable(tf.truncated_normal([784,200], stddev=0.1), name="W1")
b1 = tf.Variable(tf.zeros([200]), name="b1")
h1 = tf.nn.sigmoid(tf.matmul(x_, W1) + b1)
W2 = tf.Variable(tf.truncated_normal([200,10], stddev=0.1), name="W2")
b2 = tf.Variable(tf.zeros([10]), name="b2")
y = tf.nn.softmax(tf.matmul(h1, W2) + b2)

cross_entropy = -tf.reduce_sum(y_ * tf.log(y))

optimizer = tf.train.GradientDescentOptimizer(0.01)
trainer = optimizer.minimize(cross_entropy)

init = tf.initialize_all_variables()
```

## Session

```python
sess = tf.Session()
sess.run(init)
```

## 建立成效評估用的Computational Graph

```python
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

## Saver

```python
saver = tf.train.Saver()
```

## 訓練

```python
patience = 50
best_accurarcy = 0
i = 0
while True:
    i += 1
    batch_xs, batch_ys = mnist.train.next_batch(100)
    sess.run(trainer,  feed_dict={x_: batch_xs, y_: batch_ys})
    if i%20 == 0:
        train_accurarcy = sess.run(accuracy, feed_dict={x_: batch_xs, y_: batch_ys})
        valid_accurarcy = sess.run(accuracy, feed_dict={x_: mnist.validation.images, y_: mnist.validation.labels})
        print "%s, train: %s, valid: %s" %(i, train_accurarcy, valid_accurarcy)
        if valid_accurarcy > best_accurarcy:
            patience = 50
            best_accurarcy = valid_accurarcy
            print "save model"
            saver.save(sess, "model_multilayer_perceptron.ckpt")
        else:
            patience -= 1
            if patience == 0:
                print "early stop"
                break 
```

```python
saver.restore(sess, "model_multilayer_perceptron.ckpt")
valid_accurarcy = sess.run(accuracy, feed_dict={x_: mnist.validation.images, y_: mnist.validation.labels})
test_accurarcy = sess.run(accuracy, feed_dict={x_: mnist.test.images, y_: mnist.test.labels})
print "valid:%s, test:%s"%(valid_accurarcy,test_accurarcy)
# valid:0.98 ,test:0.9779
```
